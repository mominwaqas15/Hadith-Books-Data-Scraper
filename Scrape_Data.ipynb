```python
book_urls = ['https://sunnah.com/bukhari', 'https://sunnah.com/muslim', 'https://sunnah.com/nasai',
             'https://sunnah.com/abudawud', 'https://sunnah.com/tirmidhi', 'https://sunnah.com/ibnmajah',
             'https://sunnah.com/malik', 'https://sunnah.com/ahmad', 'https://sunnah.com/darimi']

book_names = []
chapter_numbers = []
english_titles = []
arabic_titles = []

# Iterate over each book URL
for url in book_urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    book_name = url.split('/')[-1]

    book_titles = soup.find_all('div', class_='book_title')

    for title in book_titles:
        chapter_numbers.append(title.find('div', class_='book_number').text.strip())
        english_titles.append(title.find('div', class_='english').text.strip())
        arabic_titles.append(title.find('div', class_='arabic').text.strip())
        book_names.append(book_name)

data = {'Book Name': book_names, 'Chapter Number': chapter_numbers, 'English Title': english_titles, 'Arabic Title': arabic_titles}
df_book_names = pd.DataFrame(data)

df_book_names.to_csv("Book Name.csv")

df_book_names.head(20)
```

## tirmidhi

```python
base_url = 'https://sunnah.com/tirmidhi/'

data_tirmidhi = []

# Define the range of chapter numbers
start_chapter = 1
end_chapter = 49

for chapter in range(start_chapter, end_chapter + 1):
    # Construct the URL for the current chapter
    url = base_url + str(chapter)

    # Send a GET request to the URL
    response = requests.get(url)

    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the necessary elements and extract the data
    chapter_div = soup.find('div', class_='chapter')
    if chapter_div is not None:
        english_chapter = chapter_div.find('div', class_='englishchapter')
        english_chapter_text = english_chapter.text.strip() if english_chapter is not None else ''

        arabic_chapter = chapter_div.find('div', class_='arabicchapter arabic')
        arabic_chapter_text = arabic_chapter.text.strip() if arabic_chapter is not None else ''

        hadith_container = soup.find('div', class_='actualHadithContainer')
        if hadith_container is not None:
            hadith_text = hadith_container.find('div', class_='english_hadith_full')
            hadith_text = hadith_text.text.strip() if hadith_text is not None else ''

            hadith_arabic_text = hadith_container.find('div', class_='arabic_hadith_full')
            hadith_arabic_text = hadith_arabic_text.text.strip() if hadith_arabic_text is not None else ''

            hadith_reference = hadith_container.find('div', class_='hadith_reference_sticky')
            hadith_reference = hadith_reference.text.strip() if hadith_reference is not None else ''

            data_tirmidhi.append({
                'Chapter': chapter,
                'English Chapter': english_chapter_text,
                'Arabic Chapter': arabic_chapter_text,
                'Hadith': hadith_text,
                'Arabic Hadith': hadith_arabic_text,
                'Hadith Reference': hadith_reference
            })
        else:
            print('No Hadith Container found for Chapter', chapter)
    else:
        print('Chapter', chapter, 'not found')
    print()

df_tirmidhi = pd.DataFrame(data_tirmidhi)

df_tirmidhi.head(20)
```
